```python
df = pd.read_csv("avocado.csv")
df = df.copy()[df['type']=='organic']

df["Date"] = pd.to_datetime(df["Date"])

df.sort_values(by="Date", ascending=True, inplace=True)
df.head()
```

```python
graph_df=pd.DataFrame()
for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df['region']==region]
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df["{}_price25ma".format(region)] = region_df["AveragePrice"].rolling(25).mean()

    if graph_df.empty:
        graph_df = region_df[["{}_price25ma".format(region)]]  # note the double square brackets!
    else:
        graph_df = graph_df.join(region_df["{}_price25ma".format(region)])
        

        
graph_df.tail()
```

```python
graph_df.plot(figsize=(8,5),legend=False)
```
